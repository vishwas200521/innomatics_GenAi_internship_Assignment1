```python
def smart_parking(capacity, logs):
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1
        if current_parked > peak_usage:
            peak_usage = current_parked

    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)

    if current_parked > capacity:
        print("Alert: Parking Exceeded Capacity!")
    else:
        print("Parking Status: Available")
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    
    base_speed = 20 
    preparation_time = 20 
    travel_time = (distance / base_speed) * 60
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 20
    else:
        traffic_delay = 0
    
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Rainy":
        weather_delay = 10
    elif weather == "Stormy":
        weather_delay = 20
    else:
        weather_delay = 0
    
    total_time = preparation_time + travel_time + traffic_delay + weather_delay
    
    print("Estimated Delivery Time:", int(total_time), "minutes")
estimate_delivery_time(8, "High", "Rainy")
```

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 40:
        status = "Moderate"
    else:
        status = "Low Occupancy"
    if occupancy >= 90:
        suggestion = "Suggestion: Open additional shows"
    else:
        suggestion = "Suggestion: No additional shows needed"
    print("Total Seats:", total_seats)
    print("Booked Seats:", booked_count)
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print(suggestion)
total_seats = 200


booked_seats = [1] * 150

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):
    
    total = 0
    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)
    if average_cpu < 50:
        status = "Normal"
    elif 50 <= average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    print("Average CPU Load:", int(average_cpu), "%")
    print("Server Status:", status)
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def monitor_resource_usage(resource_usage):
    
    overused_resources = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(usage)
```

```python
def manage_registrations(capacity, total_registrations):
    
    confirmed = 0
    waitlisted = 0
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1
    if confirmed >= capacity:
        status = "Closed"
    else:
        status = "Open"
    
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
manage_registrations(100, 105)   
```
