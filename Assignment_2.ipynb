```python
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_employee = max(employees, key=employees.get)
print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

```python
num = []
n = int(input("How many numbers in a list: "))
for i in range(n):
    number = int(input("Enter the number: "))
    num.append(number)
print("Maximum:", max(num))
print("Minimum:", min(num))
```

```python
price = []
n = int(input("How many products? "))

for i in range(n):
    prices = int(input("Enter price: "))
    price.append(prices)

count =0
for i in price :
    if i>1000:
        count=count+1
print("products above 1000 is :",count)
```

```python
attendence=["p","p","a","p","a"]
count=0
for i in attendence:
    if "p"==i:
        count=count+1
length=len(attendence)
percentage=(count/length) * 100
print("Attendence percentage :",percentage)
```

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_numbers)
```

```python
text = input("Enter a text:")
frequency = {}
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1
print(frequency)
```
