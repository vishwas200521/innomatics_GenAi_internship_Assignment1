```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print(" Recharge amount must be at least ₹50.")
        return False
    if amount not in valid_plans:
        print(" Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False
    return True
while True:
    amount = int(input("Enter recharge amount: ₹"))
    
    if validate_recharge(amount):
        print("Recharge successful!")
        break   
    else:
        print("Please try again.\n")
```

```python
def check_inventory(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert  (Stock = {quantity})")
        else:
            print(f"{product}: Stock OK (Stock = {quantity})")
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Wheat": 18,
    "Salt": 8,
    "Oil": 20
}
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    if average >= 50:
        return f"Average: {average:.2f}  Pass "
    else:
        return f"Average: {average:.2f}  Fail "
student_marks = [60, 45, 70, 50, 55]
result = process_result(student_marks)
print(result)
```

```python
def calculate_fare(distance, is_peak):
    base_fare = 50
    per_km_rate = 12
    fare = base_fare + (distance * per_km_rate)
    if is_peak:
        fare += fare * 0.25   
    return fare
while True:
    distance = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()
    is_peak = True if peak_input == "yes" else False
    total_fare = calculate_fare(distance, is_peak)
    print(f"Total Fare: ₹{total_fare:.2f}")
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry == "no":
        print("Thank you for using Cab Fare Estimator")
        break
```

```python
def check_eligibility(attendance_list):
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100
    
    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% Eligible "
    else:
        return f"Attendance: {percentage:.2f}%  Not Eligible"
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_eligibility(attendance)
print(result)
```

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    has_digit = False
    has_special = False
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must contain at least one digit and one special character)"
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
