```python
class SmartLight:
    def __init__(self, name):
        self.light_name = name
        self.is_on = False  

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        if self.is_on:
            print(f"{self.light_name} is ON")
        else:
            print(f"{self.light_name} is OFF")
name = input("Enter Light Name: ")
action = input("Enter Action (ON/OFF): ")

light = SmartLight(name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\n----- Employee ID Card -----")
        print(f"Employee Name : {self.name}")
        print(f"Employee ID   : {self.emp_id}")
        print(f"Department    : {self.department}")
    
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")
employee = Employee(name, emp_id, department)
employee.display_id_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")
name = input("Contact Name: ")
phone = input("Phone Number: ")


contact = Contact(name, phone)
contact.display_contact()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
name = input("Product Name: ")
price = input("Price: ")
product = Product(name, price)
product.print_price_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = float(rating)

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")
name = input("Movie Name: ")
rating = input("Rating: ")
movie = Movie(name, rating)
movie.display_details()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")
name = input("Customer Name: ")
address = input("Address: ")
delivery = Delivery(name, address)
delivery.print_delivery_details()
```
