```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
highest_score = max(employees.values())
top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)
print("Top Performers Eligible for Bonus:", 
      ", ".join(top_performers), 
      f"(Score: {highest_score})")
```

```python
query = input("Enter search query: ")
query = query.lower()
query = query.replace(".", "")
query = query.replace(",", "")
query = query.replace("!", "")
query = query.replace("?", "")
words = query.split()
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] = frequency[word] + 1
    else:
        frequency[word] = 1
print("Keywords searched more than once:")
for word in frequency:
    if frequency[word] > 1:
        print(word, ":", frequency[word])
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for i in range(len(sensor_readings)):
    reading = sensor_readings[i]
    # Check if reading is even
    if reading % 2 == 0:
        valid_readings.append((i, reading))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in emails:  
    parts = email.split("@")
    domain = parts[1] 
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1
total_emails = len(emails)
for domain in domain_count:
    percentage = (domain_count[domain] / total_emails) * 100
    print(domain + ":", int(percentage), "%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total = 0
for value in sales:
    total = total + value
average = total / len(sales)
threshold = average + (0.30 * average)
for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
frequency = {}
for user in user_ids:
    if user in frequency:
        frequency[user] += 1
    else:
        frequency[user] = 1
print("Duplicate User IDs:")
for user, count in frequency.items():
    if count > 1:
        print(f"{user} → {count} times")
```
